```python
import tensorflow.contrib.keras.api.keras as K
import pandas as pd
```

```python
SEED = 1234
N_DATA = 1000
DATA_DIR = "data_keras_claim_line/"
DATASETS_DIR = "~/Work/machine-learning/"
```

```python
claim_line_clean = pd.read_csv(DATASETS_DIR + "CLAIM_LINE_CLEAN.csv")
claim_line_clean = claim_line_clean.sample(n=N_DATA, random_state=SEED)
```

```python
model = K.models.Sequential()
model.add(K.layers.Embedding(input_dim=1, output_dim=1000, input_length=694))
model.compile(optimizer='adam', loss='binary_crossentropy')
```

```python
K.utils.plot_model(model, show_shapes=True)
import IPython
IPython.display.Image(filename='model.png')
```

```python
model.fit(claim_line_clean.values, claim_line_clean.values, epochs=1, verbose=1)
```
