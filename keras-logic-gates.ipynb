```python
import tensorflow.contrib.keras.api.keras as K
from tensorflow.contrib.keras.api.keras import backend as B

import numpy as np

#np.set_printoptions(formatter={'float': lambda x: "{0:0.5f}".format(x)})
```

```python
FILENAME = "keras-logic-gates"

L_XOR = 0
L_OR  = 1
L_AND = 2

x_data = np.array([
    [0, 0, L_XOR],
    [0, 1, L_XOR],
    [1, 0, L_XOR],
    [1, 1, L_XOR],
    [0, 0, L_OR],
    [0, 1, L_OR],
    [1, 0, L_OR],
    [1, 1, L_OR],
    [0, 0, L_AND],
    [0, 1, L_AND],
    [1, 0, L_AND],
    [1, 1, L_AND],
])

y_data = np.array([
    #XOR
    [0],
    [1],
    [1],
    [0],
    #OR
    [0],
    [1],
    [1],
    [1],
    #AND
    [0],
    [0],
    [0],
    [1],
])

# hyperparameters
n_input = 3
n_output = 1
n_hidden = 20
learning_rate = 0.1
epochs = 1000
batch_size = 12

model = K.models.Sequential()
model.add(K.layers.Dense(n_hidden, activation=B.relu, input_shape=(n_input,)))
model.add(K.layers.Dense(n_output, activation=B.sigmoid))
```

```python
sgd = K.optimizers.SGD(lr=learning_rate)
model.compile(optimizer=sgd, loss='mean_squared_error', metrics=['acc'])
```

```python
callback_tensorboard = K.callbacks.TensorBoard(log_dir="./data-" + FILENAME, write_images=True)

model.fit(x_data, y_data, epochs=epochs, batch_size=batch_size, verbose=1, callbacks=[callback_tensorboard])
```

```python
model.evaluate(x_data, y_data)
```

```python
#XOR
res = model.predict(x_data[0:4], verbose=1)
print(res)
print()
print(res.round())
```

```python
#OR
res = model.predict(x_data[4:8], verbose=1)
print(res)
print()
print(res.round())
```

```python
#AND
res = model.predict(x_data[8:12], verbose=1)
print(res)
print()
print(res.round())
```
